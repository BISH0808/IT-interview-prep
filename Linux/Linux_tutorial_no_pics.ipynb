## Навигация по файлам и директориям

###Основнвые команды

* cd  - сменить директорию
* ls – вывести содержимое директории
* pwd – вывести путь где мы сейчас
* ~    - сокращение нашей Home директории
* /    - коренная директория Linux
* ..    - директория которая выше 
* .     - директория где мы сейчас

cd / - перейти в самую коренную директорию
ls - показать директории
ls -l -показать директории (long format)
ls -la - показать все директории, в том числе скрытые директории и файлы (all)
cd .. - вернуться на одну директорию вверх
pwd - посмотреть где я 
cd ../../ -вернуться на 2 директории выше
cd - вернуться в свою домашнюю директорию
cd ~ - также вернуться в свою домашнюю директорию
cd ~/Documents/ - перейти сразу в Документы
